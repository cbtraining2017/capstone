```python
import requests
import pandas as pd
import numpy as np
```

```python
toronto_url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
download = requests.get(toronto_url)
download.status_code
```

```python
page=download.text
```

```python
toronto_list=pd.read_html(page,header=0,attrs={"class":"wikitable sortable"})[0]
```

```python
toronto_list.head()
```

```python
not_assinged = toronto_list[toronto_list['Borough']== 'Not assigned'].index
```

```python
toronto_list.drop(not_assinged, inplace=True)
```

```python
toronto_list.head()
```

```python
grouped = toronto_list.groupby(['Postcode','Borough'], sort = False).agg(lambda x: ','.join(x))
```
